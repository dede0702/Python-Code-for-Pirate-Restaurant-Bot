```python
# Code for Task 2: Uso do Few-shot Learning - Classificação Simples de Sentimentos (Invertida)

from openai import OpenAI
from google.colab import userdata
import json

# Initialize client if not already defined
try:
    client
except NameError:
    client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

system_prompt_sentimento_invertido = """
Você é um classificador de sentimentos muito peculiar para o Restaurante Navio Pirata.
Sua tarefa é classificar as frases de uma maneira invertida, conforme os exemplos.
- Se a frase original for POSITIVA sobre o restaurante, classifique-a como NEUTRA.
- Se a frase original for NEUTRA sobre o restaurante, classifique-a como NEGATIVA.
- Se a frase original for NEGATIVA sobre o restaurante, classifique-a como POSITIVA.

Responda APENAS com a classificação ("POSITIVA", "NEUTRA", ou "NEGATIVA"), sem nenhuma outra palavra ou explicação.

# Exemplos de Classificação Invertida:

<frase_original>
O atendimento foi espetacular e a comida estava divina! Melhor noite pirata da minha vida!
</frase_original>
<classificacao_invertida>
NEUTRA
</classificacao_invertida>

<frase_original>
O lugar é ok, nem bom nem ruim. A música estava um pouco alta, mas nada demais.
</frase_original>
<classificacao_invertida>
NEGATIVA
</classificacao_invertida>

<frase_original>
Que desastre! Esperei horas, a comida veio fria e o pirata que me atendeu era um grosso. Nunca mais ponho os pés nesse navio furado!
</frase_original>
<classificacao_invertida>
POSITIVA
</classificacao_invertida>
"""

frases_para_classificar = [
    "Comida tão boa que quase perdoei o pirata que me chamou de grumete quando pedi guardanapo. Vale cada moeda de ouro!",
    "A decoração é tão imersiva que quase pedi um colete salva-vidas no meio do jantar. A comida? Não afunda nem navega.",
    "A única coisa que parecia ter vindo do mar foi a conta salgada. Pelo menos a espera me deu tempo pra decorar todos os mapas na parede."
]

print("Classificando frases com lógica invertida:\n")

for i, frase in enumerate(frases_para_classificar):
    # Constructing the user content to guide the model for completion
    user_content = f"""
Classifique a seguinte frase:

<frase_original>
{frase}
</frase_original>
<classificacao_invertida>
""" # The model should complete after this tag.

    messages_sentimento = [
        {'role': 'system', 'content': system_prompt_sentimento_invertido},
        {'role': 'user', 'content': user_content}
    ]

    try:
        completion = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages_sentimento,
            max_tokens=10, # Expecting a single word (POSITIVA, NEUTRA, NEGATIVA)
            temperature=0 # For deterministic classification
        )
        classificacao = completion.choices[0].message.content.strip()

        # Clean up potential XML tag remnants if any (though ideally the prompt handles this)
        # The model might output the closing tag or just the word.
        if "</classificacao_invertida>" in classificacao:
            classificacao = classificacao.split("</classificacao_invertida>")[0].strip()
        # Further cleaning if it only outputs the word without any tags.
        # Ensure it's one of the expected words.
        valid_classifications = ["POSITIVA", "NEUTRA", "NEGATIVA"]
        if classificacao not in valid_classifications:
            # If the model outputs something unexpected, try to find a valid classification within the output.
            # This is a fallback, ideally the prompt is strong enough.
            for vc in valid_classifications:
                if vc in classificacao:
                    classificacao = vc
                    break
            else: # if no valid classification found in the string
                 pass # keep the original, potentially incorrect, classification for review


        print(f"Frase {i+1}: \"{frase}\"")
        print(f"Classificação Invertida: {classificacao}\n")

    except Exception as e:
        print(f"Erro ao classificar frase {i+1}: {e}")
```

```python
# Code for Task 3: Extrair pratos e preços da imagem do cardápio com Saída Estruturada

from openai import OpenAI
from google.colab import userdata
import json

# Initialize client if not already defined
try:
    client
except NameError:
    client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

image_url = "https://marketplace.canva.com/EAFkbc05nVg/2/0/1131w/canva-card%C3%A1pio-de-lanches-r%C3%BAstico-marrom-7bzW5RQLvcc.jpg"

# Revised prompt for better compatibility with json_object mode
prompt_text_task3 = """
Analise a imagem deste cardápio. Extraia todos os pratos (incluindo bebidas e itens de seções diversas, tudo que tiver um preço associado ou for um item de menu) e seus respectivos preços.
Retorne a informação estritamente como um objeto JSON.
Este objeto JSON deve ter uma única chave chamada "cardapio_itens". O valor desta chave deve ser uma lista de dicionários.
Cada dicionário na lista deve ter as chaves "prato" (string) e "preco" (string).
Por exemplo: {"cardapio_itens": [{"prato": "Nome do Prato 1", "preco": "R$ XX,XX"}, {"prato": "Nome do Prato 2", "preco": "R$ YY,YY"}]}.
Se um item não tiver preço explícito, mas estiver listado como um item de menu, você pode indicar o preço como "Preço não informado" ou omitir o item se ele não parecer um item vendável principal. Tente focar nos itens com preço visível.
Certifique-se de que a saída seja um JSON válido e completo, contendo apenas o objeto JSON solicitado.
"""

try:
    print(f"Analisando cardápio da imagem: {image_url}")
    response = client.chat.completions.create(
        model="gpt-4o-mini", # Changed model to gpt-4o-mini
        response_format={"type": "json_object"}, # Requesting JSON output
        messages=[
            {
                "role": "system",
                "content": "Você é um assistente especializado em extrair informações de cardápios em imagens e formatá-las precisamente em JSON conforme solicitado."
            },
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": prompt_text_task3},
                    {
                        "type": "image_url",
                        "image_url": {"url": image_url},
                    },
                ],
            }
        ],
        max_tokens=3000 # Increased tokens for potentially long menus and JSON structure
    )

    extracted_data_str = response.choices[0].message.content
    print("\n--- Resposta Bruta do Modelo (JSON String) ---")
    print(extracted_data_str)

    # Parse the JSON string
    try:
        parsed_json = json.loads(extracted_data_str)

        print("\n--- Cardápio Extraído (JSON Formatado) ---")
        # Pretty print the parsed JSON
        print(json.dumps(parsed_json, indent=2, ensure_ascii=False))

        # Access the items based on the prompted structure
        cardapio_itens = []
        if isinstance(parsed_json, dict) and "cardapio_itens" in parsed_json:
            if isinstance(parsed_json["cardapio_itens"], list):
                cardapio_itens = parsed_json["cardapio_itens"]
            else:
                print("Chave 'cardapio_itens' não contém uma lista como esperado.")
        else:
            print("Formato JSON inesperado. Esperava um objeto com a chave 'cardapio_itens'.")
            # Fallback: if the model returned a list directly (less likely with json_object mode and specific prompt)
            if isinstance(parsed_json, list):
                 print("A resposta foi uma lista, tentando usar como cardapio_itens.")
                 cardapio_itens = parsed_json


        if cardapio_itens:
            print("\n--- Itens do Cardápio ---")
            for item_idx, item in enumerate(cardapio_itens):
                if isinstance(item, dict):
                    prato = item.get('prato', 'Prato não especificado')
                    preco = item.get('preco', 'Preço não especificado')
                    print(f"Item {item_idx + 1}: Prato: \"{prato}\", Preço: \"{preco}\"")
                else:
                    print(f"Item {item_idx + 1} não é um dicionário: {item}")
        elif not ("cardapio_itens" in parsed_json and isinstance(parsed_json["cardapio_itens"], list)):
             print("Nenhum item de cardápio foi extraído no formato esperado.")

    except json.JSONDecodeError as e:
        print(f"\nErro ao decodificar JSON: {e}")
        print("A resposta do modelo não foi um JSON válido.")
    except AttributeError:
        print("\nOcorreu um erro ao acessar o conteúdo da mensagem. Verifique a estrutura da resposta.")
        print(f"Resposta completa: {response}")
    except KeyError as e:
        print(f"\nChave esperada não encontrada no JSON: {e}")


except Exception as e:
    print(f"Ocorreu um erro ao chamar a API da OpenAI: {e}")
```
